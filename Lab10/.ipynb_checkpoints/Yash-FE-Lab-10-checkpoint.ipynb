```python
import numpy as np
import matplotlib.pyplot as plt
from pylab import rcParams
import math
rcParams['figure.figsize'] = 10, 10
```

```python
S=100.0
mu = 0.1
r = 0.05
sigma = 0.2
delT = 1.0/365
K=90
Y = 0
T=180
m=100
delT = 1.0/365
S0 = 100


def getStockPricesWithVarianceReduction():
    Splus = np.ndarray(T)
    Sminus = np.ndarray(T)
    Splus[0] = 100.0
    Sminus[0] = 100.0
    for i in range(1,T):
        W =  math.sqrt(delT)*np.random.randn()
        Splus[i] = Splus[i-1]*np.exp(sigma*W + (r-0.5*(sigma**2))*delT)
        Sminus[i] = Sminus[i-1]*np.exp(sigma*(-W) + (r-0.5*(sigma**2))*delT)
    return Splus,Sminus

def getStockPrices():
    S = np.ndarray(T)
    S[0] = 100.0
    for i in range(1,T):
        W =  math.sqrt(delT)*np.random.randn()
        S[i] = S[i-1]*np.exp(sigma*W + (r-0.5*(sigma**2))*delT)
    return S,S

def calculateOptionPrice(m,withReduction=False):
    payoffSum = 0
    optionPrice = 0
    for i in range(m):
        if withReduction ==True:
            Splus,Sminus = getStockPricesWithVarianceReduction()
        else:
            Splus,Sminus = getStockPrices()

        Yplus = np.mean(Splus)
        Yminus = np.mean(Sminus)
        payoff = 0.5*(max(0,Yplus-K) + max(0,Yminus-K))
        payoffSum = payoffSum + payoff 
    return np.exp(-r*delT*T)*(payoffSum/m)



def Ques12(K):
    a1 = []
    a2 = []
    for i in range(100,10000,100):
        print(i)
        a1.append(calculateOptionPrice(i))
        a2.append(calculateOptionPrice(i,withReduction=True))
```
