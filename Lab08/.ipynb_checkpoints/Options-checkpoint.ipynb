```python
from nsepy import get_history
from datetime import date
import pandas
```

```python
nifty_opt = get_history(symbol="SBIN", start=date(2014,1,1),
                        end=date(2019,3,29),
                        index=True,
                        option_type='CE',
                        strike_price=300,
                        expiry_date=date(2019,3,29))
```

```python
nifty_opt
```

```python
stock_opt = get_history(symbol="SBIN",
start=date(2015,1,1),
end=date(2015,1,10),
option_type="CE",
strike_price=300,
expiry_date=date(2015,1,29))
```

```python
stock_opt
```
